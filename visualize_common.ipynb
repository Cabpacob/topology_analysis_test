```python

import os
import ast
import networkx
from python_graphs import control_flow, control_flow_graphviz, program_graph_graphviz
from IPython.display import Image, display
from python_graphs import program_graph
from graphviz import Source
import tempfile
```

```python
def visualize(source):
    graph = control_flow.get_control_flow_graph(source)

    file = tempfile.NamedTemporaryFile(suffix = '.png')
    control_flow_graphviz.render(graph, path=file.name)

    plt = Image(file.name)
    display(plt)
```

```python
visualize(
"""
a = 10
b = 20
c = a + b
"""
)
```

```python
visualize(
"""
if a > 0:
    print(1)
elif a < 0:
    print(-1)
else:
    print(0)
"""
)
```

```python
visualize(
"""
def sum(a, b):
    return a + b
"""
)
```
